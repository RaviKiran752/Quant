```python
import requests
import pandas as pd

# Initialization function to store API details
def initialize_upstox(api_key, api_secret, code, headers):
    return {
        'api_key': api_key,
        'api_secret': api_secret,
        'code': code,
        'headers': headers
    }

# Get instrument data
def get_instruments():
    furl = 'https://assets.upstox.com/market-quote/instruments/exchange/complete.csv.gz'
    symbolZ = pd.read_csv(furl)
    return symbolZ

# Logout function
def logout(headers):
    url_logout = "https://api.upstox.com/v2/logout"
    response3 = requests.request("DELETE", url_logout, headers=headers)
    print(response3.text)
    return response3

# Get user profile
def get_profile(headers):
    url_profile = "https://api.upstox.com/v2/user/profile"
    response4 = requests.request("GET", url_profile, headers=headers)
    return response4

# Get funds and margin details
def get_funds_margin(headers):
    url = "https://api.upstox.com/v2/user/get-funds-and-margin"
    response5 = requests.request("GET", url, headers=headers)
    return response5

# Get brokerage charges
def get_brokerage(headers, params):
    url = "https://api.upstox.com/v2/charges/brokerage"
    response6 = requests.request("GET", url, headers=headers, params=params)
    return response6

# Get margin details for an instrument
def get_margins(access_token):
    url = "https://api.upstox.com/v2/charges/margin"
    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {access_token}",
        "Content-Type": "application/json"
    }
    data = {
        "instruments": [
            {
                "instrument_key": "NSE_EQ|INE669E01016",
                "quantity": 1,
                "transaction_type": "BUY",
                "product": "D"
            }
        ]
    }
    response = requests.post(url, headers=headers, json=data)
    print(response.status_code)
    print(response.json())
    return response

# Get historical data
def get_historical_data(instrument_key,duration, start_date, end_date, access_token):

    base_url = 'https://api.upstox.com/v2/historical-candle'
    # instrument_key = 'NSE_EQ%7CINE848E01016'  # Example instrument key
    # duration = '1minute'  # Example duration
    # start_date = '2023-11-13'  # Start date
    # end_date = '2023-11-12'  # End date

    url = f'{base_url}/{instrument_key}/{duration}/{start_date}/{end_date}'
    # url = f'https://api.upstox.com/v2/historical-candle/NSE_EQ%7CINE848E01016/{duration}/{start_date}/{end_date}'
    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {access_token}",
        "Content-Type": "application/json"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        print(response.json())
    else:
        print(f"Error: {response.status_code} - {response.text}")
    return response

# Get long-term holdings
def get_holdings(headers):
    url = 'https://api.upstox.com/v2/portfolio/long-term-holdings'
    response = requests.get(url, headers=headers)
    return response

# Get intraday data
def get_intraday_data(headers, duration):
    url = f'https://api.upstox.com/v2/historical-candle/intraday/NSE_EQ%7CINE848E01016/{duration}'
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        print(response.json())
    else:
        print(f"Error: {response.status_code} - {response.text}")
    return response

# Get live market data
def get_live_data(headers, instrument_key):
    url = f'https://api.upstox.com/v2/market-quote/quotes?instrument_key={instrument_key}'
    response = requests.get(url, headers=headers)
    return response

# Get option chain data
def get_option_chain(headers):
    url = 'https://api.upstox.com/v2/option/chain'
    params = {
        'instrument_key': 'NSE_INDEX|Nifty 50',
        'expiry_date': '2024-12-19'
    }
    response = requests.get(url, params=params, headers=headers)
    return response
```
